## **1. What is the difference between "Love" and "love" in NLP?**
In NLP, "Love" and "love" can be treated as two different tokens if case normalization is not applied. That means the model may think they are separate words even though they usually carry the same meaning. Converting text to lowercase helps reduce this inconsistency.


# **2. What happens if stopwords are not removed?**

If stopwords are not removed, the text may contain many common words like is, the, am, was, and, which usually do not add much meaning. This can increase noise, make the feature space larger, and reduce model efficiency. However, whether to remove them depends on the task.


# **3. Two real-world scenarios where removing stopwords can be harmful**

Sentiment analysis: Words like not, no, and never are very important.
Example: "I am not happy" becomes "happy" if stopwords are removed carelessly, which changes the meaning completely.

Question answering / search systems: Stopwords may affect intent.
Example: "to be or not to be" or "flights from Mumbai to Delhi" lose important structure if stopwords are removed.

# **4. Difference between stemming and lemmatization**

Stemming cuts words to a rough base form using simple rules.
Example: playing, played, player may become play or even incomplete forms like pla in some cases.

Lemmatization converts a word to its correct dictionary form using language knowledge.
Example: better becomes good, running becomes run.


# **2: Advanced Preprocessing Function**

```python
import re
from collections import Counter

def preprocess_text(text):
    """
    Cleans text and returns tokens.

    Features:
    - remove numbers
    - remove extra spaces
    - handle repeated characters
    - convert to lowercase
    - remove very short tokens (length <= 2)
      except: keep 'no', 'not'
    - remove URLs and email-like patterns
    - remove punctuation/emojis/special characters
    """

    if text is None:
        return []

    if not isinstance(text, str):
        text = str(text)

    # lowercase
    text = text.lower()

    # remove urls
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)

    # remove email-like patterns
    text = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', ' ', text)

    # remove numbers
    text = re.sub(r'\d+', ' ', text)

    # keep only alphabets, apostrophes, and spaces
    text = re.sub(r"[^a-z\s']", ' ', text)

    # handle repeated characters: 3 or more -> 1
    # soooo -> so, baaad -> bad, looooved -> loved
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    if not text:
        return []

    tokens = text.split()

    # remove short tokens except meaningful exceptions
    exceptions = {"no", "not"}
    tokens = [token for token in tokens if len(token) > 2 or token in exceptions]

    return tokens

    # List of sentences
text = ["I have 2 dogs","This is   good","soooo goooood!!!","WOW!!! This is GREAT!!!","Visit http://example.com now"  ]

# Process each sentence
for sentence in text:
    result = preprocess_text(sentence)
    print("Original:", sentence)
    print("Tokens:", result)
    print("Cleaned:", " ".join(result))
    print("-" * 40)
```

# **3: Stress Testing**

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

def cleaned_sentence_from_tokens(tokens):
    return " ".join(tokens)

for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)
    clean_sentence = cleaned_sentence_from_tokens(tokens)

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 50)
```

# **4: Token Analytics**

```python
def token_analytics(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = round(sum(len(token) for token in tokens) / total_tokens, 2) if total_tokens > 0 else 0
    return total_tokens, unique_tokens, avg_token_length

for i, sentence in enumerate(test_sentences, 1):
    tokens = preprocess_text(sentence)
    total, unique, avg_len = token_analytics(tokens)

    print(f"Sentence {i}:")
    print("Total number of tokens :", total)
    print("Unique tokens          :", unique)
    print("Average token length   :", avg_len)
    print("-" * 50)
```

# **5: Frequency Analysis**

```python
all_tokens = []

for sentence in test_sentences:
    all_tokens.extend(preprocess_text(sentence))

word_freq = Counter(all_tokens)

top_10_most_frequent = word_freq.most_common(10)
least_5_frequent = word_freq.most_common()[:-6:-1]

print("Top 10 most frequent words:")
print(top_10_most_frequent)

print("\nTop 5 least frequent words:")
print(least_5_frequent)
```

# **6:Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
